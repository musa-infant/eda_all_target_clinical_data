# Esplorazione dei dati clinici relativi a ALL
https://www.cbioportal.org/study/clinicalData?id=all_phase2_target_2018_pub

Di seguito le colonne del dataset:
- **Study ID**: Identificativo dello studio o del protocollo di riferimento.
- **Patient ID**: Codice univoco assegnato al paziente.
- **Sample ID**: Codice univoco assegnato al campione biologico (es. sangue o tessuto) prelevato dal paziente.
- **Diagnosis Age**: Età del paziente al momento della diagnosi (in anni).
- **Diagnosis Age (days)**: Età del paziente al momento della diagnosi (in giorni).
- **Alternative Therapy Given**: Indica se sono state somministrate terapie alternative rispetto allo standard.
- **Other Alternative Therapy Given**: Dettagli su altre terapie alternative eventualmente somministrate.
- **Analysis Cohort**: Gruppo di analisi o sottostudio di appartenenza del paziente (ad esempio per protocolli clinici specifici).
- **BCR ABL1 Status**: Stato di fusione BCR-ABL1, tipicamente indicato in leucemie (positivo/negativo). Come specificato [qui](https://www.ogt.com/media/zqzfttb4/lph007-ce-it-v013-00.pdf), nei bambini la fusione BCR-ABL1 (da traslocazione t(9;22)) può comparire in una parte dei casi di leucemia linfoblastica acuta (LLA) e ne peggiora la prognosi. L’identificazione precoce di questa anomalia è fondamentale per definire il rischio e stabilire la terapia più adatta.
- **Bone Marrow Blasts Day 15**: Percentuale (o conteggio) di blasti nel midollo osseo al giorno 15 dall’inizio del trattamento.
- **Bone Marrow Blasts Day 29**: Percentuale (o conteggio) di blasti nel midollo osseo al giorno 29 dall’inizio del trattamento.
- **Bone Marrow Blasts Day 43**: Percentuale (o conteggio) di blasti nel midollo osseo al giorno 43 dall’inizio del trattamento.
- **Bone Marrow Blasts Day 8**: Percentuale (o conteggio) di blasti nel midollo osseo al giorno 8 dall’inizio del trattamento.
- **Bone Marrow Site of Relapse**: Indica se la recidiva è stata localizzata nel midollo osseo.
- **Cancer Type**: Tipologia di cancro (macro-categoria, ad es. ALL, AML, ecc.).
- **Cancer Type Detailed**: Tipologia di cancro dettagliata (sottotipo o classificazione più specifica).
- **Cell of tumor origin**: Tipologia di cellula di origine tumorale (es. linfoblasti B, T, ecc.).
- **CNS Site of Relapse**: Indica se la recidiva ha interessato il sistema nervoso centrale (CNS).
- **CNS Status**: Stato del coinvolgimento del sistema nervoso centrale (ad es. CNS1, CNS2, CNS3).
- **Congenital Abnormality**: Presenza di anomalia/e congenita/e nel paziente.
- **Death from Initial Pathologic Diagnosis Date**: Data di morte (se applicabile) calcolata a partire dalla diagnosi iniziale.
- **Time To Event (days)**: Intervallo di tempo (in giorni) fino all’evento clinico di interesse (ad es. recidiva, morte, follow-up).
- **DNA Index**: Misura del contenuto di DNA relativo (utilizzata specialmente in leucemia pediatrica).
- **Ethnicity Category**: Categoria etnica del paziente.
- **ETV6-RUNX1 Fusion Status**: Presenza o assenza della fusione genica ETV6-RUNX1 (comune in alcune leucemie).
- **First Event**: Primo evento avverso o clinico significativo registrato (es. recidiva, morte, progressione).
- **Fraction Genome Altered**: Percentuale del genoma che presenta alterazioni (dato tipico di analisi di sequenziamento/citogenetica).
- **Karyotype**: Risultato dell’analisi citogenetica standard (caratterizzazione dei cromosomi).
- **MLL Status**: Presenza di riarrangiamenti del gene MLL (ora KMT2A), tipici di alcune leucemie.
- **Molecular Subtype**: Sottotipo molecolare della patologia (es. B-ALL iperdiploide, T-ALL, ecc.).
- **MRD Percentage at end of Consolidation**: Percentuale di Malattia Minima Residua (MRD) alla fine della fase di consolidamento.
- **MRD Percentage at end of Consolidation Sensitivity**: Sensibilità del test MRD alla fine della consolidazione (limite di rilevazione).
- **MRD Percentage Day 29**: Percentuale di MRD al giorno 29.
- **MRD Percentage Day 29 Sensitivity**: Sensibilità del test MRD al giorno 29.
- **MRD Percentage Day 43**: Percentuale di MRD al giorno 43.
- **MRD Percentage Day 43 Sensitivity**: Sensibilità del test MRD al giorno 43.
- **MRD Percentage Day 8**: Percentuale di MRD al giorno 8.
- **MRD Percentage Day 8 Sensitivity**: Sensibilità del test MRD al giorno 8.
- **Mutation Count**: Numero complessivo di mutazioni rilevate nel campione.
- **Oncotree Code**: Codice standardizzato (OncoTree) che classifica il tipo di tumore.
- **Overall Survival Days**: Sopravvivenza complessiva in giorni (calcolata dalla diagnosi o da un punto di inizio definito).
- **Overall Survival (Months)**: Sopravvivenza complessiva in mesi.
- **Overall Survival Status**: Stato alla data di ultimo follow-up (ad es. “Alive” / “Deceased”).
- **Other Site of Relapse**: Altre sedi di recidiva (diverse da midollo, CNS, testicoli, ecc.).
- **Protocol**: Protocollo di trattamento clinico a cui è stato sottoposto il paziente.
- **Race Category**: Categoria di razza / gruppo etnico del paziente.
- **Number of Samples Per Patient**: Numero di campioni totali disponibili per lo stesso paziente.
- **Sex**: Sesso del paziente (M/F/Altro).
- **Somatic Status**: Indica se la variante genetica riscontrata è di tipo somatico (tumore) o germinale.
- **TCF3-PBX1 Status**: Presenza o assenza della fusione TCF3-PBX1 (traslocazione rilevante nella leucemia).
- **Testes Site of Relapse**: Indica il coinvolgimento testicolare in caso di recidiva.
- **Testicular Involvement**: Specifica se la leucemia ha coinvolto i testicoli (alla diagnosi o successivamente).
- **TMB (nonsynonymous)**: Tumor Mutational Burden (carico mutazionale) totale per mutazioni nonsynonymous.
- **Trisomy 4_10**: Indica la presenza di trisomia dei cromosomi 4 e 10, caratteristica di alcune leucemie pediatriche.
- **WBC**: Conta dei globuli bianchi (White Blood Cells) al momento della diagnosi o in un esame specifico.

### Overview di tutte la variabili

```python
def overview_df(df, top_n=5):
    total_rows = df.shape[0]
    
    for col in df.columns:
        print(f"Column: {col}")
        
        # Missing values
        missing = df[col].isnull().sum()
        missing_perc = (missing / total_rows) * 100
        print(f"  Missing: {missing} ({missing_perc:.2f}%)")
        
        # Distinct values
        distinct = df[col].nunique(dropna=True)
        distinct_perc = (distinct / total_rows) * 100
        print(f"  Distinct: {distinct} ({distinct_perc:.2f}%)")
        
        # Valori più frequenti
        print("  Most frequent values:")
        top = df[col].value_counts(dropna=False).head(top_n)
        for value, count in top.items():
            perc = (count / total_rows) * 100
            print(f"    {value}: {count} ({perc:.2f}%)")
        
        # Stats se la colonna è numerica
        if pd.api.types.is_numeric_dtype(df[col]):
            print("  Stats:")
            print(f"    Min: {df[col].min()}")
            print(f"    Max: {df[col].max()}")
            print(f"    Mean: {df[col].mean():.2f}")
            print(f"    Median: {df[col].median()}")
        
        print("-" * 40)


overview_df(df, top_n=5)
```

```python
import pandas as pd

def overview_df(df, top_n=5):
    total_rows = df.shape[0]
    rows = []

    for col in df.columns:
        # Calcolo dei missing values
        missing = df[col].isnull().sum()
        missing_perc = (missing / total_rows) * 100
        missing_str = f"{missing} ({missing_perc:.2f}%)"
        
        # Calcolo dei valori distinti
        distinct = df[col].nunique(dropna=True)
        distinct_perc = (distinct / total_rows) * 100
        distinct_str = f"{distinct} ({distinct_perc:.2f}%)"
        
        # Se la colonna è numerica, usa describe() per ottenere tutte le statistiche
        if pd.api.types.is_numeric_dtype(df[col]):
            descr = df[col].describe()
            stats_str_list = []
            for stat, value in descr.items():
                if stat == "count":
                    # Per count, mostriamo l'intero
                    stats_str_list.append(f"{stat}: {int(value)}")
                else:
                    stats_str_list.append(f"{stat}: {value:.2f}")
            stats_str = ", ".join(stats_str_list)
            mfv_str = ""  # Non valorizzato per colonne numeriche
        else:
            # Per colonne categoriche, calcola i valori più frequenti
            top = df[col].value_counts(dropna=False).head(top_n)
            mfv_list = [
                f"{value}: {count} ({(count/total_rows*100):.2f}%)" 
                for value, count in top.items()
            ]
            mfv_str = "; ".join(mfv_list)
            stats_str = ""  # Non valorizzato per colonne non numeriche
        
        rows.append({
            "Column": col,
            "Missing Values": missing_str,
            "Distinct Values": distinct_str,
            "Most Frequent Values": mfv_str,
            "Stats": stats_str
        })
        
    return pd.DataFrame(rows)


df_overview = overview_df(df, top_n=3)
df_overview.to_csv('overview.csv', sep=';', index=False)
```

## Righe duplicate
Dall'analisi visiva del dataset sembra che le righe per stesso Patient (colonna Patient ID) siano duplicati, eccetto per pochissime colonne. Verifichiamo quali colonne non presentano valori duplicati.

```python
# Inizializziamo un dizionario per memorizzare le colonne problematiche
non_unique_columns = {}

for col in [col for col in df.columns]:
    # Per ogni colonna, raggruppiamo per Patient ID e contiamo i valori unici
    unique_counts = df.groupby('Patient ID')[col].nunique(dropna=False)
    max_unique = unique_counts.max()  # valore massimo di valori unici per Patient ID in questa colonna
    if max_unique > 1:
        non_unique_columns[col] = max_unique

if non_unique_columns:
    print("\nLe colonne che presentano valori differenti per stesso Patient ID:")
    for col, max_val in non_unique_columns.items():
        print(f"- {col}")
```

```python
non_unique_columns
list(non_unique_columns.keys())
```

```python
# Verifichiamo il numero di righe distinte escludendo le colonne precedentemente trovate

group_keys = [col for col in df.columns if col not in list(non_unique_columns.keys())]

# Raggruppa il DataFrame per queste colonne
df_grouped = df.groupby(group_keys, dropna=False).size().reset_index(name='counts')

print("Numero di gruppi:", df_grouped.shape[0])

# Conta il numero di gruppi ottenuti
num_groups = df_grouped.shape[0]

# Conta il numero di 'Patient ID' unici nel DataFrame originale
num_unique_patients = df['Patient ID'].nunique()

print(f"Numero di gruppi (raggruppando per tutte le colonne tranne {list(non_unique_columns.keys())}):", num_groups)
print("Numero di 'Patient ID' unici:", num_unique_patients)

if num_groups == num_unique_patients:
    print("Verifica OK: il numero di gruppi corrisponde al numero di 'Patient ID' unici.")
else:
    print("Attenzione: il numero di gruppi NON corrisponde al numero di 'Patient ID' unici.")
```

Considerando che queste caratteristiche hanno un'elevata percentuale di missing (Oltre 60%, fino ad oltre il 90%), possiamo eliminare queste colonne e di conseguenza aggregare i dati per eliminare i duplicati, in modo da avere una riga per ogni paziente

```python
df_temp = df.copy()

# Rimozione delle colonne non uniche
df_temp.drop(columns=non_unique_columns.keys(), inplace=True)

# Eliminazione  dei duplicati
df_temp.drop_duplicates(inplace=True)
```

### Analizzare l’associazione tra variabili categoriche
Per capire se due variabili categoriche sono associate (es. CNS Site of Relapse vs. Bone Marrow Site of Relapse), usiamo una tabella di contingenza:

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df['CNS Site of Relapse'], df['Bone Marrow Site of Relapse'])
print(table)

chi2, p, dof, expected = chi2_contingency(table)
print(f"Chi2 = {chi2}, p-value = {p}")
```
